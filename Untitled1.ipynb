```python
from itertools import cycle

import numpy as np
import pylab as plt
colors = cycle(plt.mpl.colors.TABLEAU_COLORS)

from bornagain.target import Molecule, Atoms
```

```python
mol = Molecule(pdbFilePath='.local/ext/bornagain/examples/data/pdb/1jb0.pdb') # PS1
monomers = mol.get_monomers()




#======== PLOTTING CODE (BORING)=================

plt.figure( figsize=(16,16))
ax = plt.gca()

# plot each monomer and label it with its center of mass
stride=240 # every 240th atom plot we shall
for i,mono in enumerate(monomers):
    c = colors.next()
    com = mono.xyz.mean(0) # center of mass
    com_lab = ", \n".join( map( lambda x: '%.1f'%x, com)) # str label
    com_lab = "mono %d:\n%s"%(i, com_lab)

    
    plt.scatter(mono.xyz[::stride,0], mono.xyz[::stride,1], s=14, c=c)
    ax.text(com[0], com[1], com_lab, fontsize=20, backgroundcolor='w', alpha=0.5, )
    
# add the xtal directions
def plot_ab(ax):
    ax.arrow( 0,0,mol.lat.a[0], mol.lat.a[1], head_width=10, head_length=10, ec='r', fc='r', lw=3)
    ax.arrow(0,0,mol.lat.b[0], mol.lat.b[1], head_width=10, head_length=10, ec='b', fc='b', lw=3)
    ax.text( mol.lat.a[0], 20, 'a', color='r', fontsize=30)
    ax.text(mol.lat.b[0]-20, mol.lat.b[1]+20,'b', color='b', fontsize=30)

plot_ab(ax)
ax.set_xlim(-300,300)
ax.set_ylim(-300,300)
plt.show()
#=========================================


# Notice that 3 monomers have average Z coordinate of 74, and 3 have average Z coordinate of 156.6
# this is how we can identify the trimers.. 
```

```python
# We would like to assemble each trimer together within a unit cell
# From the above, to form the trimer at Z=74.0, we should shift mono2 by one unit of a, and mono4 by one unit of a and b
# This is best done in fractional coordinates

monomers[2] = mol.shift(monomers[2], 1,0,0) # shift along a once
monomers[4] = mol.shift( monomers[4], 1,1,0) # shift along a,b each once

# likewise we can shift mono5 along b once and mono3 along a,b each once
monomers[3] = mol.shift( monomers[3],1,1,0)
monomers[5] = mol.shift(monomers[5],0,1,0)
```

```python
# now group the monomers into trimers we must
trimer1 = Atoms.aggregate( [monomers[0], monomers[2], monomers[4]])
trimer2 = Atoms.aggregate( [monomers[1], monomers[3], monomers[5]])
```

```python
# ================ PLOT CODE ===============
#                   -_-

# now that we grouped the monomers, lets replot
plt.figure( figsize=(16,16))

# replot each monomer with the shifts
c1,c2 = colors.next(), colors.next()

plt.scatter(trimer1.xyz[::stride,0], trimer1.xyz[::stride,1], s=40, c=c2, label='trimer1')
plt.scatter(trimer2.xyz[::stride,0], trimer2.xyz[::stride,1], s=40, c=c1, label='trimer2')
plt.legend()

# add the xtal directions
plot_ab(plt.gca())

plt.xlim(-200,300)
plt.ylim(-50,300)
plt.xlabel(r'x ($\AA$)')
plt.ylabel(r'y ($\AA$)')
plt.show()
# ====================================================
```

```python
# Make the PS1 lattice
mol.lat.assemble(n_unit=20, spherical=False) # makes the lattice for PS1 (origin is at 0)
print (mol.lat.vecs)
```

```python
# view lattice along z-direction
def plot_lattice(subplots, lat, **kwargs): # convenience function
    fig,(ax1, ax2, ax3) = subplots
    
    ax1.scatter(lat[:,0], lat[:,1], **kwargs)
    ax1.set_xlabel(r'x ($\AA$)')
    ax1.set_ylabel(r'y ($\AA$)')

    ax2.scatter(lat[:,0], lat[:,2], **kwargs)
    ax2.set_xlabel(r'x ($\AA$)')
    ax2.set_ylabel(r'z ($\AA$)')

    ax3.scatter(lat[:,1], lat[:,2], **kwargs)
    ax3.set_xlabel(r'y ($\AA$)')
    ax3.set_ylabel(r'z ($\AA$)')
    #plt.subplots_adjust(hspace=0.2)
    
subplots = plt.subplots(nrows=3, ncols=1, figsize=(10,30))
plot_lattice(subplots,  mol.lat.vecs, **{'s':20, 'c':'b'})
plt.show()
```

```python
# lets make 2 lattices, one for each trimer1, and another for each trimer2
lat1 = mol.lat.vecs + trimer1.xyz.mean(0) # add trimer center of mass
lat2 = mol.lat.vecs + trimer2.xyz.mean(0)

# now lets look at the lattices

subplots = plt.subplots(nrows=3, ncols=1, figsize=(10,30))
plot_lattice(subplots,  lat1, **{'s':20, 'c':'b'})
plot_lattice(subplots,  lat2, **{'s':20, 'c':'Darkorange'})
plt.show()
```

```python
# simulate the scattering we must
from bornagain.simulate import  ThornAgain
from bornagain.detector import SimpleDetector

D = SimpleDetector(n_pixels=2000, wavelen=6.5, pixsize=0.00007, detdist=0.2, name="yoda") # somewhat realistic ...
print("Using this detector:")
print(D)

# subtract the center of mass from each trimer because we shifted the lattices 
# according to their center of masses...
print("Initializing trimer 1 simulator")
%time Tmol1 = ThornAgain(D.Q, trimer1.xyz-trimer1.xyz.mean(0), trimer1.Z) # init sims

print("\nInitializing trimer 2 simulator")
%time Tmol2 = ThornAgain(D.Q, trimer2.xyz-trimer2.xyz.mean(0), trimer2.Z)

print("\nInitializing trimer 2 simulator")
%time Tlat1 = ThornAgain(D.Q, lat1)

print("\nInitializing trimer 2 simulator")
%time Tlat2 = ThornAgain(D.Q, lat2)

```

```python
print("Trimer 1 amplitudes...")
Amol1 = Tmol1.run(rand_rot=False)
print("\nTrimer 2 amplitudes...")
Amol2 = Tmol2.run(rand_rot=False)
print("\nLattice 1 amplitudes...")
Alat1 = Tlat1.run(rand_rot=False)
print("\nLattice 2 amplitudes...")
Alat2 = Tlat2.run(rand_rot=False)
```

```python
image = D.readout_finite(Amol1*Alat1 + Amol2*Alat2, qmin=0.004, qmax=0.4, flux=1e22)
```

```python

def plot_image(image):
    plt.figure(figsize=(16,16))

    qx1, qy1 , _ = D.Q.min(0)
    qx2, qy2, _ = D.Q.max(0)
    extent = ( qx1, qx2, qy1, qy2)

    plt.imshow(image+1, norm=plt.mpl.colors.LogNorm(), cmap='gnuplot', extent=extent)

    cbar = plt.colorbar()
    cbar.ax.set_title('photon counts') #, rotation=270, labelpad=12)

    plt.xlabel(r'$q_x\,\,\AA^{-1}$')
    plt.ylabel(r'$q_y\,\,\AA^{-1}$')
    plt.show()
    
plot_image(image)
```

```python
# Now do a random orientation
print("Trimer 1 amplitudes...")
Amol1 = Tmol1.run(rand_rot=True)
rot_mat = Tmol1.rot_mat

print("\nTrimer 2 amplitudes...")
Amol2 = Tmol2.run(force_rot_mat=rot_mat)

print("\nLattice 1 amplitudes...")
Alat1 = Tlat1.run(force_rot_mat=rot_mat)

print("\nLattice 2 amplitudes...")
Alat2 = Tlat2.run(force_rot_mat=rot_mat)

imageR = D.readout_finite(Amol1*Alat1 + Amol2*Alat2, qmin=0.004, qmax=0.4, flux=1e22)
```

```python
plot_image(imageR)
```
